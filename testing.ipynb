```python
# Bert

from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

Preprocessing of the utterance by removing the stop words + lemmatizer using SpaCy.

```python
TARGET_DOMAINS = ["restaurant", "hotel"]
UNKNOWN_DOMAIN = "unknown"

X = []
y = []

# Accessing a single dialogue
def add_dialogue_to_json(dialogue, json):

    # Iterating through the turns to get the utterances
    for i in range( 0, len(dialogue["utterance"]), 2 ):
        utterance = dialogue["utterance"][i]
        domains = dialogue["frames"][i]["service"]

        domains = [UNKNOWN_DOMAIN if domain not in TARGET_DOMAINS else domain for domain in domains]
        if len(domains) == 0:
            domains = [UNKNOWN_DOMAIN]

        # utterance, vector = preprocess_utterance(utterance)
        print(utterance)
        vector = model.encode(utterance)
        if vector.size == 0: # Some utterances are only stopwords. We hate those
          continue

        # json.append(
        #     {
        #         'domains': domains,
        #         'utterance': utterance
        #     }
        # )

        domains.sort()
        y.append(" ".join(domains))
        X.append(vector)

    return json

json = []

for i, item in enumerate(dataset["train"]):

    if not any(set(item['services']).intersection(['hotel', 'restaurant'])):
        continue
    add_dialogue_to_json(item["turns"], json)
    # if i == 100:
    #   break

```

```python
X = np.array(X)
y = np.array(y)
np.save('X_bert.npy', X)
np.save('y_bert.npy', y)
```
